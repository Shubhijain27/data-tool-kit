 Explain the following terms :

 The difference in np.array, np.asarray and np.asanyarray

 The difference between Deep copy and shallow copyX

np.array(), np.asarray(), and np.asanyarray():

np.array(): Creates a NumPy array from an input object. It attempts to convert the input to a NumPy array, and if successful, returns a new array with the specified dtype and optional copy and subok parameters. If the input cannot be converted, it raises a ValueError.
np.asarray(): Similar to np.array(), but it always returns a NumPy array, even if the input is already an array. It also attempts to preserve the original array’s dtype and layout (C- or Fortran-contiguous).
np.asanyarray(): Converts the input to a NumPy array, but allows it to retain its original dtype and layout if it’s already a subclass of ndarray (e.g., np.matrix). If the input is not an ndarray subclass, it attempts to convert it to a regular ndarray.


Deep Copy vs Shallow Copy:

Shallow Copy: A copy of the original object that references the same memory locations as the original. This means that if the original object contains mutable elements (like lists or dictionaries), modifying those elements in the copied object will also affect the original. NumPy’s np.copy() function performs a shallow copy.
Deep Copy: A copy of the original object that recursively creates new copies of all its mutable elements, ensuring that changes to the copied object do not affect the original. NumPy does not have a built-in deep copy function, but you can use Python’s copy module or third-party libraries like dill or pickle to achieve deep copying.


```python
import numpy as np

a = np.arange(5)
b = np.array(a)  # creates a copy of a
print(a is b)
```

```python
import numpy as np

m = np.matrix([[1, 2], [3, 4]])
b = np.asanyarray(m)  # returns m (no copy)
print(m is b)  # True

a = np.arange(5)
b = np.asanyarray(a)  # returns a (no copy)
print(a is b)  # True
```

```python
import numpy as np

# creating array
a = np.array([ 2, 3, 4, 5, 6])
print("Original array : ",a)

# assigning value to np.array
np_array = np.array(a)
a[3] = 0
print("np.array Array : ",np_array)

# assigning value to np.asarray
np_array = np.asarray(a)
print("np.asarray Array : ",np_array)
```

```python
import copy

original_list = [[1, 2], [3, 4]]
shallow_copied_list = copy.copy(original_list)

print(shallow_copied_list)

# Modify the copied list
shallow_copied_list[0][0] = 10

print(original_list)
```

```python
import copy

original_list = [[1, 2], [3, 4]]
deep_copied_list = copy.deepcopy(original_list)

print(deep_copied_list)

# Modify the copied list
deep_copied_list[0][0] = 10

print(original_list)
```

```python
#includes all the parts
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set seed for reproducibility
np.random.seed(0)

# Create NumPy array with 1000 samples from a normal distribution
mean, std_dev = 0, 1  # mean and standard deviation
data = np.random.normal(mean, std_dev, 1000)

# Create histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Overlay normal distribution's PDF
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r-', lw=2, label='Normal Distribution PDF')

# Set plot properties
plt.title('Histogram with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.legend()

# Display plot
plt.show()
```

Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
rng = np.random.RandomState(0)
x = rng.uniform(-10, 10, 500)
y = rng.uniform(-10, 10, 500)

# Define quadrant-based color mapping
colors = ['red' if x[i] * y[i] > 0 else 'blue' for i in range(len(x))]

# Create the scatter plot
sns.scatterplot(x=x, y=y, hue=colors, palette=['red', 'blue'], legend="brief")

# Set title, labels, and axis limits
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Show the plot
plt.show()
```

ques 19

 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'8


```python



import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

np.random.seed(0)
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

df = pd.DataFrame({'Category': categories, 'Value': values})

source = ColumnDataSource(data=df)

p = figure(title="Random Categorical Bar Chart",
           x_axis_label='Category',
           y_axis_label='Value',
           x_range=categories,
           height=400,
           width=600,
           toolbar_location="above")

p.vbar(x='Category', top='Value', width=0.9,
       line_color='white',
       fill_color='blue',
       source=source)

hover = HoverTool(tooltips=[
    ("Category", "@Category"),
    ("Value", "@Value"),
])

p.add_tools(hover)

show(p)
```

Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate random dataset
np.random.seed(0)  # for reproducibility
x = np.random.uniform(0, 10, 50)  # x-axis values
y = np.random.normal(0, 2, 50)  # y-axis values

# Create a Pandas DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Create the line plot
fig = px.line(df, x='x', y='y', title='Simple Line Plot')

# Customize axis labels and font
fig.update_layout(
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    font=dict(family='Arial', size=12)
)

# Show the plot
fig.show()
```

 Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(0)
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, 4)

# Create a Pandas DataFrame
df = pd.DataFrame({'Label': labels, 'Value': values})

# Create the pie chart
fig = px.pie(df, values='Value', names='Label', title='Interactive Pie Chart',
             hover_data=['Label', 'Value'], labels={'Value': 'Percentage (%)'})

# Show the plot
fig.show()
```
